```python
x=df.iloc[:,df.columns!="Outcome"]#independent data
y=df.iloc[:,df.columns=="Outcome"]#dependent data
```

## train and test dataset


```python
from sklearn.model_selection import train_test_split
```

```python
x
```

```python
xtrain, xtest, ytrain, ytest=train_test_split(x,y,test_size=0.2,random_state=23) #test20% train80%
```

## algorithm

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain.values.ravel()) #to train model
```

```python
predi_output=model.predict(xtest)#to test model
```

```python
predi_output
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(predi_output,ytest)
```
